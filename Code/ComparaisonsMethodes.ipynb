```python
import time
from FonctionsInstances import *
from RicochetRobotSolverPronfondeur import *
from RicochetRobotSolverLargeur import *
from RicochetRobotSolverA_H1 import *
from RicochetRobotSolverA_H2 import *

from matplotlib import pyplot as plt
```

```python
#TEST UNIQUEMENT LA METHODE DE RECHERCHE EN LARGEUR 

kMax = 3
nMax = 12

for n in range(8,nMax+1,2):    
    m = n #nous avons choisi de definir le max de profondeur à n (pour eviter que le programme dure trop longtemps)
    temps_moyen = []
    nbCoups_moyen = []
    for k in range(1,kMax+1,1):
        timesLargeur = []
        nbCoupsLargeur = []
        for i in range(20):
            cellules, verticaux, horizontaux = generateRandomInstances(n, k)

            SolvProfondeur = RicochetRobotSolverPronfondeur(n, k, m, cellules.copy(), verticaux, horizontaux)
            _, nbCoupsSolProf, _ = SolvProfondeur.Recherche()

            t0 = time.time()

            SolvLargeur = RicochetRobotSolverLargeur(n, k, nbCoupsSolProf, cellules.copy(), verticaux, horizontaux)
            _, nbCoup, _ = SolvLargeur.Recherche()

            t1 = time.time()

            timesLargeur.append(t1-t0)
            nbCoupsLargeur.append(nbCoup)
        temps_moyen.append(sum(timesLargeur)/20)
        nbCoups_moyen.append(sum(nbCoupsLargeur)/20)
        
    plt.plot(list(range(1,kMax+1,1)), temps_moyen, marker='o', linestyle='-', label=f"Recherche en largeur avec {k} robots")

    plt.title(f"Comparaisons des temps de recherche - methode en largeur sur 20 instances de taille {n}")
    plt.xlabel("Nombres de robots")
    plt.ylabel('Temps de recherche moyen (en seconde)')
    plt.legend()
    plt.show()

    plt.figure()
    
    plt.plot(list(range(1,kMax+1,1)), nbCoups_moyen, marker='o', linestyle='-', label=f"Recherche en largeur avec {k} robots")

    plt.title(f"Comparaisons des temps de recherche - methode en largeur sur 20 instances de taille {n}")
    plt.xlabel("Nombres de robots")
    plt.ylabel('Nombre de coups moyens des solution trouvés')
    plt.legend()
    plt.show()
```

```python
#TEST COMPARAISONS DES PERFORMANCES DES 3 METHODES SUR LES MEMES INSTANCES

kMax = 3
nMax = 12

for k in range(1,kMax+1,1):
    for n in range(8,nMax+1,2): 
        m = n #profondeur max de la methode recherche en profondeur
        
        timesLargeur = []
        nbCoupsLargeur = []
        
        timesAlgoA_H1 = []
        nbCoupsAlgoA_H1 = []
        
        timesAlgoA_H2 = []
        nbCoupsAlgoA_H2 = []
        
        for i in range(20):
            cellules,verticaux,horizontaux=generateRandomInstances(n,k)
            
            #Si j'ai bien compris rech. prof uniquement pour obtenir la borne nbCoupsSolProf, alors on prend pas son temps
            SolvProfondeur = RicochetRobotSolverPronfondeur(n,k,m,cellules.copy(),verticaux,horizontaux)
            _, nbCoupsSolProf, _ = SolvProfondeur.Recherche()

            # // CALCUL DU TEMPS  ET NB DE COUPS POUR RECHERCHE EN LARGEUR
            t0 = time.time()

            SolvLargeur = RicochetRobotSolverLargeur(n,k,nbCoupsSolProf,cellules.copy(),verticaux,horizontaux)
            _, nbCoup, _ = SolvLargeur.Recherche()

            t1 = time.time()
            
            timesLargeur.append(t1-t0)
            nbCoupsLargeur.append(nbCoup)
            
            # // CALCUL DU TEMPS  ET NB DE COUPS POUR RECHERCHE AVEC ALGORITHME A* HEURISTIQUE N°1
            t0 = time.time()

            SolvAlgoA_H1 = RicochetRobotSolverA_H1(n,k,nbCoupsSolProf,cellules.copy(),verticaux,horizontaux)
            _, nbCoup, _ = SolvAlgoA_H1.Recherche()

            t1 = time.time()
            
            timesAlgoA_H1.append(t1-t0)
            nbCoupsAlgoA_H1.append(nbCoup)
            
            # // CALCUL DU TEMPS  ET NB DE COUPS POUR RECHERCHE AVEC ALGORITHME A* HEURISTIQUE N°2
            t0 = time.time()

            SolvAlgoA_H2 = RicochetRobotSolverA_H2(n,k,nbCoupsSolProf,cellules.copy(),verticaux,horizontaux)
            _, nbCoup, _ = SolvAlgoA_H2.Recherche()

            t1 = time.time()
            
            timesAlgoA_H2.append(t1-t0)
            nbCoupsAlgoA_H2.append(nbCoup)

        # // COMPARAISON DU TEMPS POUR LES DIFFERENTES METHODES DE RECHERCHES
        plt.figure()

        plt.plot(list(range(1,21)), timesLargeur, marker='o', linestyle='-',label="Recherche en largeur")
        plt.plot(list(range(1,21)), timesAlgoA_H1, marker='+', linestyle='-',label="Recherche avec Algo A* heuristique 1")
        plt.plot(list(range(1,21)), timesAlgoA_H2, marker='x', linestyle='-',label="Recherche avec Algo A* heuristique 2")

        plt.title(f"Temps de recherche des differentes methodes sur 20 instances de taille {n} avec {k} robots")
        plt.xlabel('i-ème instance')
        plt.ylabel('Temps de recherche (en seconde)')
        plt.xticks(range(1,21))

        plt.legend()

        plt.show()

        print(f"Le temps moyen de l'execution avec recherche en largeur des 20 instances de taille {n} avec {k} robots est de ", sum(timesLargeur)/20, " secondes.")
        print(f"Le temps moyen de l'execution avec recherche ALGO A* heuristique 1 des 20 instances de taille {n} avec {k} robots est de ", sum(timesAlgoA_H1)/20, " secondes.")
        print(f"Le temps moyen de l'execution avec recherche ALGO A* heuristique 2 des 20 instances de taille {n} avec {k} robots est de ", sum(timesAlgoA_H2)/20, " secondes.")

        # // COMPARAISON DU NOMBRE DE COUPS POUR LES DIFFERENTES METHODES DE RECHERCHES
        plt.figure()

        plt.plot(list(range(1,21)), nbCoupsLargeur, marker='o', linestyle='-',label="Recherche en largeur")
        plt.plot(list(range(1,21)), nbCoupsAlgoA_H1, marker='+', linestyle='-',label="Recherche avec Algo A* heuristique 1")
        plt.plot(list(range(1,21)), nbCoupsAlgoA_H2, marker='x', linestyle='-',label="Recherche avec Algo A* heuristique 2")

        plt.title(f"Nombre de coups des differentes methodes sur 20 instances de taille {n} avec {k} robots")
        plt.xlabel('i-ème instance')
        plt.ylabel('Nombre de coups de la solution trouvé')
        plt.xticks(range(1,21))

        plt.legend()

        plt.show()


        print(f"Le nombre de coups moyen de l'execution de la recherche en largeur des 20 instances de taille {n} avec {k} robots est de ", sum(nbCoupsLargeur)/20, " coups.")
        print(f"Le nombre de coups moyen de l'execution de la recherche ALGO A* heuristique 1 des 20 instances de taille {n} avec {k} robots est de ", sum(nbCoupsAlgoA_H1)/20, " coups.")
        print(f"Le nombre de coups moyen de l'execution de la recherche ALGO A* heuristique 2 des 20 instances de taille {n} avec {k} robots est de ", sum(nbCoupsAlgoA_H2)/20, " coups.")
```
